# Matplotlib Library

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x= np.arange(0,20)
```

```python
y= np.arange(20,40)
```

```python
plt.scatter(x,y, c='g')
```

```python
plt.scatter(x, y, s=None )
```

```python
plt.xlabel('x-axis')
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
## Bar plot

x = [3,8,10] 
y = [10,16,9]  

x2 = [8,9,11] 
y2 = [1,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 10)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);
```

```python
# Data to plot
labels = 'Python', 'C++', 'Java'
sizes = [215, 130, 210]
colors = ['gold', 'yellowgreen', 'lightskyblue']
explode = (0.4, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
